```python
questions = {
    "ENVIRONMENT" : [
        "how many people or person or communities impacted",
        "how much area impacted",
        "how much area of land affected",
        "How much pollution is emitted into the air, water, or land",
        "How much waste is generated",
        "What are the specific health issues faced",
        "were plants or forests damaged",
        "how many animals were injured",
        "was ecosystem or habitat affected",
        "How much loss incurred financially"],

    "SOCIAL" : [
        "How many people were affected?",
        "How much fine is paid",
        "How many suicides",
        "Were there any deaths or torture or rapes",
        "How many people were injured",
        "Were any rights violated",
        "What is the company being accused of",
        "was property damaged or destroyed",
        "Were there any injuries or illness or disabilities"],

    "GOVERNANCE" : [
        "were there any frauds or bankruptcy",
        "how much bribe was given",
        "was government impacted",
        "are there incidents of corruption",
        "how many people were affected",
        "which countries were affected",
        "were there incidents of money laundering",
        "How much fine or loss was incurred"]
}
```

```python
def processArticle(article, pillar):
  question = None
  if pillar == "ENVIRONMENT":
    question = questions["ENVIRONMENT"]
  elif pillar == "SOCIAL":
    question = questions["SOCIAL"]
  elif pillar == "GOVERNANCE":
    question == questions["GOVERNANCE"]
  answers = []
  for x in question:
    answer = get_answer(x, article)
    answers.append(answer)
  return answers
```

```python
import torch
def get_answer(question, context):
  inputs = tokenizer(question, context, return_tensors="pt")
  with torch.no_grad():
    outputs = model(**inputs)
  answer_start_index = outputs.start_logits.argmax()
  answer_end_index = outputs.end_logits.argmax()
  predict_answer_tokens = inputs.input_ids[0, answer_start_index: answer_end_index + 1]
  answer=tokenizer.decode(predict_answer_tokens, skip_special_tokens=True)
  return answer
```
